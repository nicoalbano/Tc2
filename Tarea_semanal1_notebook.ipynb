* Quedandonos una transferencia de este estilo:

\begin{equation}
H(s) = \frac{s-\frac{R2}{R1*R3*C}}{s+\frac{1}{R3*C}}.

\begin{equation}
H(s) = \frac{s-k*{\omega_0^2}}{s+\omega_0^2}.
\end{equation}

* Parametrizando con:
\begin{equation}
\omega_0^2=1
\end{equation}

\begin{equation}
k=\frac{R2}{R1}
\end{equation}
* Nos quedaria:

\begin{equation}
H(s) = \frac{s-k}{s+1}.
\end{equation}

<u>La naturaleza del filtro entonces va a depender de la relacion de R2 con R1</u>

# <u>Conclusiones:</u> 

*La naturaleza de fitrado va a depender de la relacion que se encuentre entre:*
\begin{equation}
k=\frac{R2}{R1}
\end{equation} 

Cuando k sea menor a uno nos encontraremos un filtro **pasa-alto**

Cuando k sea igual a uno nos encontraremos un filtro **pasa-todo**

Cuando k sea mayor a uno nos encontraremos un filtro **pasa-bajo**

<u>Entonces podriamos obtener diferentes filtros con solo variar las resistencias.</u>
