### Parte B

1) Un Detector VNA es un instrumento de medicin utilizado para caracterizar y analizar las propiedades de redes elctricas y componentes. Su funcin principal es medir los parametros s como la reflexin (S11, S22) y la transmisin (S21, S12) de seales a travs de una red o dispositivo bajo prueba. Un VNA es fundamental en cualquier trabajo que requiera mediciones precisas de las propiedades de seal en el dominio de RF, ya sea en investigacin, diseo de dispositivos o pruebas de calidad.

2) Comparado con un DSO podemos ver:

Ancho de banda: el ancho de banda de un VNA recorre desde 1Mhz a los 50Ghz o mas lo que es crucial para el analisis de microondas mientras que un DSO va de 20Mhz a 1Ghz

Rango Dinmico: Los VNA tienen un rango dinmico muy alto, generalmente de 70 dB a 120 dB, lo que permite medir seales de muy bajo nivel (como prdidas muy pequeas) con gran precisin. Esto es fundamental en la medicin de parmetros S. Los osciloscopios tienen un rango dinmico que puede variar entre 40 dB y 100 dB, dependiendo del modelo y de la configuracin de la seal. El rango dinmico de un DSO es adecuado para visualizar seales de baja y alta amplitud, pero no alcanza el mismo nivel de precisin y detalle que un VNA en mediciones de seales dbiles o de muy alta frecuencia.

Impedancia de Entrada: VNA: La impedancia de entrada tpica de un VNA es 50 ohmios, que es el estndar en las aplicaciones de RF y microondas. Esto es importante para garantizar la correcta adaptacin de impedancia entre el VNA y los dispositivos bajo prueba (DUT), como antenas, cables, filtros, etc. La impedancia de entrada de un osciloscopio digital tambin suele ser 50 ohmios (aunque algunos modelos tienen opciones de 1 M en modo de alta impedancia para seales de baja frecuencia), lo que es til para mediciones de seales en general. Sin embargo, el DSO no est optimizado para la adaptacin precisa de impedancia en redes de RF como lo est un VNA.

3) Algunos dispositivos VNA son capaces de hacer una transformada de Fourier inversa de los parmetros S (como S11, S21, etc.) para obtener una representacin en el dominio del tiempo. Esto es particularmente til para analizar pulsos o seales transitorias y para evaluar el comportamiento de los dispositivos en trminos de retardo o dispersiones temporales.Otra medicion en tiempo es de TDR o TDT (Time Domain Transmission), que permiten observar cmo se refleja una seal cuando viaja a travs de una lnea de transmisin, cable o cualquier otro dispositivo.Estas mediciones proporcionan informacin detallada sobre la impedancia de la lnea a lo largo de su longitud, permitiendo detectar desajustes de impedancia o fallos en la integridad de la seal a lo largo del tiempo.Con un VNA, tambin puedes realizar un anlisis de respuesta transitoria observando cmo un dispositivo o red responde a una seal en funcin del tiempo. Esto es til para estudiar efectos como resolucin temporal, prdidas de seal o respuestas no lineales.
