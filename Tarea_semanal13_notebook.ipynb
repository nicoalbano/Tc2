### Parte B

1) Un Detector VNA es un instrumento de medición utilizado para caracterizar y analizar las propiedades de redes eléctricas y componentes. Su función principal es medir los parametros s como la reflexión (S11, S22) y la transmisión (S21, S12) de señales a través de una red o dispositivo bajo prueba. Un VNA es fundamental en cualquier trabajo que requiera mediciones precisas de las propiedades de señal en el dominio de RF, ya sea en investigación, diseño de dispositivos o pruebas de calidad.

2) Comparado con un DSO podemos ver:

Ancho de banda: el ancho de banda de un VNA recorre desde 1Mhz a los 50Ghz o mas lo que es crucial para el analisis de microondas mientras que un DSO va de 20Mhz a 1Ghz

Rango Dinámico: Los VNA tienen un rango dinámico muy alto, generalmente de 70 dB a 120 dB, lo que permite medir señales de muy bajo nivel (como pérdidas muy pequeñas) con gran precisión. Esto es fundamental en la medición de parámetros S. Los osciloscopios tienen un rango dinámico que puede variar entre 40 dB y 100 dB, dependiendo del modelo y de la configuración de la señal. El rango dinámico de un DSO es adecuado para visualizar señales de baja y alta amplitud, pero no alcanza el mismo nivel de precisión y detalle que un VNA en mediciones de señales débiles o de muy alta frecuencia.

Impedancia de Entrada: VNA: La impedancia de entrada típica de un VNA es 50 ohmios, que es el estándar en las aplicaciones de RF y microondas. Esto es importante para garantizar la correcta adaptación de impedancia entre el VNA y los dispositivos bajo prueba (DUT), como antenas, cables, filtros, etc. La impedancia de entrada de un osciloscopio digital también suele ser 50 ohmios (aunque algunos modelos tienen opciones de 1 MΩ en modo de alta impedancia para señales de baja frecuencia), lo que es útil para mediciones de señales en general. Sin embargo, el DSO no está optimizado para la adaptación precisa de impedancia en redes de RF como lo está un VNA.

3) Algunos dispositivos VNA son capaces de hacer una transformada de Fourier inversa de los parámetros S (como S11, S21, etc.) para obtener una representación en el dominio del tiempo. Esto es particularmente útil para analizar pulsos o señales transitorias y para evaluar el comportamiento de los dispositivos en términos de retardo o dispersiones temporales.Otra medicion en tiempo es de TDR o TDT (Time Domain Transmission), que permiten observar cómo se refleja una señal cuando viaja a través de una línea de transmisión, cable o cualquier otro dispositivo.Estas mediciones proporcionan información detallada sobre la impedancia de la línea a lo largo de su longitud, permitiendo detectar desajustes de impedancia o fallos en la integridad de la señal a lo largo del tiempo.Con un VNA, también puedes realizar un análisis de respuesta transitoria observando cómo un dispositivo o red responde a una señal en función del tiempo. Esto es útil para estudiar efectos como resolución temporal, pérdidas de señal o respuestas no lineales.
